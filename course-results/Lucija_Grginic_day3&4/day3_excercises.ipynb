```python

import requests
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO
```

```python
frequency_urls = {
    "small_cell_carcinoma_icdom": "https://progenetix.org/services/intervalFrequencies/?output=pgxseg&filters=pgx:icdom-80413",
}

frequency_table = {}
#download the frequency tables
for source_q, url in frequency_urls.items():
    response = requests.get(url, timeout=120)
    response.raise_for_status()
    df_freq = pd.read_csv(StringIO(response.text),sep = "\t",comment ="#")
    df_freq["source_q"]=source_q
    frequency_table[source_q] = df_freq
    print(f"Downloading freq table for {source_q}")


```

```python
frequency_table["small_cell_carcinoma_icdom"].columns
```

### Explanation:

> sample table = 1529 rows (one row is one biosample)

> freq_table= 3106 rows  (one row = one genomic bin, with a frequency of CNV loss across *all* biosamples)




Because the genome is split into many more bins than there are biosamples, the frequency table has many more rows than the sample table.

Because of this row number difference, one-to-one row allignment is not posssible. No mergning can be done because there are more bins than there are biosamples, and their row indicies represent different things.
